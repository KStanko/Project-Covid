# Coronavirus pandemic

End of 2019 marked the beginning of the global coronavirus pandemic. The first reports of infected people come from Wuhan in China. The beginning of 2020 brings a rapid increase in infections around the world. Initially, the World Health Organization decided not to recognize the epidemic as a public health emergency of international concern. WHO has previously warned that a wider outbreak is possible. Finally, on February 11, 2020 in Geneva, the name of the disease was announced by the director of this organization, Tedros Adhanom Ghebreyesus. He indicated that the official name for the disease caused by the new virus belonging to the coronavirus family would be COVID-19.
   COVID-19 is a contagious disease caused by severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2). The course of the disease varies. Most patients have mild symptoms and a good prognosis, typical initial symptoms of the disease include fever, dry cough, fatigue, and shallow breathing. Less common symptoms include: sputum production, headache, chills, hemoptysis, chest pain, diarrhea, nausea and vomiting, sore throat. The infection may also lead to pneumonia or multi-organ failure. 
Asignificant proportion of people(about 81%) undergo the infection asymptomatically or have mild symptoms resembling other upper respiratory diseases that disappear after about two weeks, while some patients may have acute (14%) or critical (5%) disease, which takes 3 to 6 weeks to heal. In critically ill patients who died, the time from symptom onset to death ranged from 2 to 8 weeks.
   Infection spreads between people mainly by airborne droplets, usually as a result of coughing or sneezing. The incubation period of the disease usually ranges from 2 to 14 days, with an average duration of 5 days. Sick people can infect 24â€“48 hours before the onset of disease symptoms.The possibility of contamination via ingestion is also considered. Some patients only have gastrointestinal symptoms, which may be due to contamination, e.g. with faeces. Basic infection prevention measures include frequent and thorough hand washing, keeping a distance physically from people with flu-like symptoms, and avoiding unwashed hands touching the face.It is recommended to cover the mouth and nose with a mask. People who suspect infection with the SARS-CoV-2 virus are advised to seek medical advice by telephone contact, and not by visiting a doctor in person.According to WHO, physical distance from other people should be at least 1 meter outdoors and indoors the more the better. 
   There is currently no approved targeted treatment for the disease. Treatment is symptomatic, supportive, and experimental. The mortality rate has been estimated at 1% to 3%. The risk of reinfection is currently being researched and discussed. Symptomatic treatment is usually aimed at managing symptoms and supporting body function. There is a large amount of independent research into treating COVID-19 with existing antiviral drugs. Hopes for an end to the epidemic carry the available vaccines to prevent infections.

```python
from sklearn.cluster import KMeans
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=40, 
                max_iter=400, 
                random_state=0)
    km.fit(x_std)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Clusters aomunt')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 3, 
            init = 'random', 
            n_init = 1, 
            max_iter = 300,
            tol = 1e-02,
            random_state = 0)

y_km = km.fit_predict(x_std)
```
